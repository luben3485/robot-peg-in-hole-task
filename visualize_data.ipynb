```python
import cv2
import numpy as np
import os
from ruamel import yaml
from  matplotlib import pyplot as plt
%matplotlib inline
    
def load_anno():
    with open(anno_file_path,'r') as f:
        anno_list = yaml.load(f,  Loader=yaml.RoundTripLoader)
    return anno_list
    
data_folder = '/Users/cmlab/data/pdc/logs_proto/insertion_xyzrot_eye_2021-07-18'
anno_file_path = os.path.join(data_folder, 'processed/peg_in_hole.yaml')
img_folder_path = os.path.join(data_folder, 'processed/images')
```

```python
anno_list = load_anno()
```

```python
import time
for idx in range(1):
    fig = plt.figure()
    rgb_img_path = os.path.join(img_folder_path, str(idx).zfill(6)+'_rgb.png')
    rgb_img = cv2.imread(rgb_img_path, cv2.IMREAD_COLOR)

    bbox_top_left_xy = tuple(anno_list[idx]['bbox_top_left_xy'])
    bbox_bottom_right_xy = tuple(anno_list[idx]['bbox_bottom_right_xy'])
    cv2.rectangle(rgb_img, bbox_top_left_xy, bbox_bottom_right_xy, (0, 0, 255), 2)
    rgb_img = rgb_img[:,:,::-1] 	# transform image to rgb
    plt.imshow(rgb_img)
    plt.show()
    
```
